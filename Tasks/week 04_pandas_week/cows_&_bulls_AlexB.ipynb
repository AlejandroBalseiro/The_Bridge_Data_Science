
- [Cows_and_Bulls]

Say the number generated by the computer is 1038. An example interaction could look like this:

  Welcome to the Cows and Bulls Game! 
  
  >Enter a number:

  >> 1234

  >>> 2 cows, 0 bulls

  >> 1253

  >>> 1 cow, 1 bull
  
  ...
Until the user guesses the number or write "STOP"

```python
import random

def check_cows(n,number_to_guest):
    ite = 0
    cows = 0
    n = str(n)
    number_to_guest = str(number_to_guest)
    for i in n:
        if i == number_to_guest[ite]:
            cows += 1
        ite += 1
    
    return cows

def check_bulls(n,number_to_guest):
    n = str(n)
    number_to_guest = str(number_to_guest)
    b = [i for i in number_to_guest if i in n]
    
    bull = len(b)
    
    return bull

number_to_guest = random.randint(1000,9999)

while True:

    n = input("Introduce an integuer number of four digits or write 'Exit'")
    try:
        n = int(n)
        cows = check_cows(n,number_to_guest)
        bulls = check_bulls(n,number_to_guest)
        if cows < 4:
            print (n)
            print("\t {c} cows, {b} bulls".format(c = cows, b = bulls-cows))
        else:
            print( n,"was the correct answer! you win!")
            break
    except:
        n = n.lower()
        if n == "exit":
            break
        else:
            print("INTRODUZCA UN NUMERO, NO UN TEXTO")
```
